## Ejercicio 1

```python
cadena ="El/DT perro/N come/V carne/N de/P la/DT carnicería/N y/C de/P la/DT nevera/N y/C canta/V el/DT la/N la/N la/N ./Fp"
```

```python
# Primero separa la cadena en palabras por espacios en blanco ' ', y luego cada palabra separa la palabra en sí de su tipo por '/'
L = []
for x in cadena.split():
  L.append(x.split('/'))
print(L)
```

```python
# Creación de lista con el tipo de palabra de cada palabra
Tipos=[]
for palabra,tipo in L:
    Tipos.append(tipo)
print ("Tipos =",Tipos) 
```

```python
# Creación diccionario {tipo de palabra: nº de veces de ese tipo de palabra}
dic = {}
for x in set(Tipos): # Eliminar repetidos
  dic[x] = Tipos.count(x)

for x in sorted(dic):
    print(x + " " + str(dic[x]))
```

## Ejercicio 2

```python
# Creación de lista con cada palabra en minúscula
Palabras=[]
for palabra,tipo in L:
    Palabras.append(palabra.lower())
print ("Palabras =",Palabras) 
```

```python
# Diccionario auxuliar con el tipo de cada palabra en minúscula y sus tipos, tantas veces como se repitan
dictaux = {}
for palabra, tipo in L:
  if palabra.lower() not in dictaux:
    dictaux[palabra.lower()] = []
  dictaux[palabra.lower()].append(tipo)
print(dictaux)
```

```python
# Diccionario donde la clave es la palabra, y el valor una lista con el nº de veces que aparece, y una lista para cada tipo de palabra y cuantas veces se repite
dict2 = {}
for palabra in set(Palabras):# Una clave por cada palabra, eliminar repetidos
  dict2[palabra] = [Palabras.count(palabra)]
  for tipo in sorted(set(dictaux.get(palabra))): # Para cada tipo de palabra (eliminando repetidos) añadir ese tipo y su recuento, por orden alfabetico
    dict2[palabra].append([tipo, dictaux.get(palabra).count(tipo)])

for x in sorted(dict2):
  string = x + ' ' + str(dict2[x][0])
  for tipo, frecuencia in dict2[x][1:]:
    string = string + ' ' + tipo + ' ' + str(frecuencia)
  print(string)
```

```python
# Diccionario bigrama: frecuencia
dict3 = {}
for bigrama in Bigramas:
  dict3[bigrama] = Bigramas.count(bigrama)
  
for x in dict3:
  print(str(x) + ' ' + str(dict3[x]))
```

## Ejercicio 4

```python
# Diccionarios con las probabilidades de las palabras y de las categorías, aprovechando las listas creadas en los ejercicios 1 y 2
ProbPalabras = {}
for palabra in set(Palabras):
  ProbPalabras[palabra] = Palabras.count(palabra)/len(Palabras)
print(ProbPalabras)

ProbTipos = {}
for tipo in set(Tipos):
  ProbTipos[tipo] = Tipos.count(tipo)/len(Tipos)
print(ProbTipos)
```

```python
def calcularProbabilidades(palabra):
  # Buscamos la palabra en el diccionario del ejercicio 2, que contenía su nº de apariciones y una lista para cada uno de sus tipos con su frecuencia
  if palabra in dict2:
    aux = dict2[palabra]
    frecuenciaPalabra = aux[0] # Primer elemento del diccionario del ejercicio 2: su frecuencia
    tipos = aux[1:] # El resto de elementos del diccionario: listas [tipo de palabra, frecuencia del tipo]
    for tipo, frecuenciaTipo in tipos:
      print('P( ' + tipo + ' | ' + palabra + ' ) = ' + str(round(frecuenciaTipo/frecuenciaPalabra, 6))) # 6 decimales
    for tipo, frecuenciaTipo in tipos:
      print('P( ' + palabra + ' | ' + tipo + ' ) = ' + str(round(ProbPalabras[palabra]*(frecuenciaTipo/frecuenciaPalabra)/ProbTipos[tipo], 6))) # Regla de Bayes
  else:
    print('Esa palabra es desconocida')
```
