# Curvas ROC

```python
fpA = []
unomenosfnA = []
for umbral in umbralesA:
  fpA.append(len([elem for elem in impostoresA if elem > umbral])/len(impostoresA))
  unomenosfnA.append(1 - len([elem for elem in clientesA if elem <= umbral])/len(clientesA))

fpB = []
unomenosfnB = []
for umbral in umbralesB:
  fpB.append(len([elem for elem in impostoresB if elem > umbral])/len(impostoresB))
  unomenosfnB.append(1 - len([elem for elem in clientesB if elem <= umbral])/len(clientesB))
```

```python
import matplotlib.pyplot as plt
plt.title('Caso A')
plt.plot([1.0] + fpA + [0.0], [1.0] + unomenosfnA + [0.0])
plt.xlabel('FP')
plt.ylabel('1 - FN')
plt.show()
```

# FP(FN = X) y umbral

```python
# Obtenemos el FP dado un FN concreto, indirectamente a través del umbral
def getFP(fn, umbrales, fps, unomenosfns):
  fn_mas_cercano = min(unomenosfns, key=lambda x:abs(x-(1-fn))) # fn en la lista más cercano para el valor de fn de entrada
  index = unomenosfns.index(fn_mas_cercano)
  fp = fps[index]
  umbral = umbrales[index]
  return fp, umbral
```

```python
getFP(0.2, umbralesA, fpA, unomenosfnA)
```

# FN(FP = X) y umbral

```python
def getFN(fp, umbrales, fps, unomenosfns):
  fp_mas_cercano = min(fps, key=lambda x:abs(x-fp)) # fn en la lista más cercano para el valor de fn de entrada
  index = fps.index(fp_mas_cercano)
  fn = 1 - unomenosfns[index]
  umbral = umbrales[index]
  return fn, umbral
```

```python
getFN(0.2, umbralesA, fpA, unomenosfnA)
```

# FP = FN y umbral

```python
def getFPequalFN(umbrales, fps, unomenosfns):
  fp = fps[0]
  fn = 1 - unomenosfns[0]
  umbral = umbrales[0]
  min_dif = 100 
  for i, x in enumerate(umbrales[1:]): # da igual que lista cojamos, las 3 son del mismo tamaño
    dif = abs(fps[i] - (1 - unomenosfns[i]))
    if (dif < min_dif):
      min_dif = dif
      fp = fps[i]
      fn = 1 - unomenosfns[i]
      umbral = umbrales[i]
  return fp, fn, umbral
```

```python
getFPequalFN(umbralesA, fpA, unomenosfnA)
```

# Area bajo la curva ROC

```python
def getAUC(clientes, impostores):
  area = 0
  for cliente in clientes:
    for impostor in impostores:
      if (cliente > impostor):
        area += 1
      elif (cliente == impostor):
        area += 0.5
      # else, cliente < impostor, area+ = 0
  area /= (len(clientes) * len(impostores))
  return area
```

```python
getAUC(clientesA, impostoresA)
```

```python
getAUC(clientesB, impostoresB)
```

# D-Prime

```python
import statistics
import math

def dprime(clientes, impostores):
  varianzaImpostores = statistics.variance(impostores)
  varianzaClientes = statistics.variance(clientes)
  mediaImpostores = statistics.mean(impostores)
  mediaClientes = statistics.mean(clientes)
  resultado = (mediaClientes - mediaImpostores) / (math.sqrt(varianzaImpostores + varianzaClientes))
  return resultado
```

```python
dprime(clientesA, impostoresA)
```

```python
dprime(clientesB, impostoresB)
```
